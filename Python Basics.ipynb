When a List is created and assigned to a variable, only the reference is passed to the variable
This means that when you assigned the variable to another variable, this happens:

````python
x = [1,2,3,4,5,6]
print(x)
y = x
print(y)
y[0] = 10
y[-1] = 99
print(y)
print(x)
````

this means, we need to use list() or slicing to assign a variable referencing a list to another variable

````python
x = [1,2,3,4,5,6]
y = list(x)
print(y)
y[0] = 10
print(x)
print(y)
z = x[:]
print(z)
````

<h2>NumPy</h2>


````python
PTS = [20,25,12,40,15,18,14,11]
TRB = [11,3,19,5,3,4,7,1]
AST = [8,8,2,5,3,1,2,7]
STL = [0,2,4,1,1,0,2,3]
BLK = [3,2,0,0,2,1,2,0]

#FTPS = 1*PTS + 1.2*TRB + 1.5*AST + 3*(STL + BLK)
#FPTS = 1 * PTS + (1.2 * TRB)
#FPTS
````

````python
import numpy as np
arrPTS = np.array(PTS)
arrTRB = np.array(TRB)
arrAST = np.array(AST)
arrSTL = np.array(STL)
arrBLK = np.array(BLK)

FPTS = 1*arrPTS + 1.2*arrTRB + 1.5*arrAST + 3*(arrSTL + arrBLK)
FPTS
````

````python
FPTS > 32
````

````python
FPTS = FPTS[FPTS > 32]
FPTS
````

<h2>Generating Data</h2>


````python
#A normal Distribution
IQ = np.round(np.random.normal(100,15,100000))
````

````python
import matplotlib.pyplot as plt

plt.hist(IQ,bins=20)
plt.show()
````

````python
mean = np.mean(IQ)
std = np.std(IQ)

standardized = (IQ - mean)/std

plt.hist(standardized,bins=20)
plt.show()
````

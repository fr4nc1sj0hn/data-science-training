<h3>Pure Textual Features</h3>

<p>
    Creating a corpus of words and counting their frequency is an extremely powerful encoding tool. AKA. bag of words technique.
</p>

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = responses["ItemText"]
```

```python
vec =  CountVectorizer()
X = vec.fit_transform(corpus) # Sparse Matrix
X
```

```python
vec.get_feature_names()
```

```python
print(vec.vocabulary_)
```

<p>
<b>SciPy</b> is a library of mathematical algorithms and convenience functions that further extend NumPy.

<i>X</i> is now a sparse matrix instead of a classical dataframe is because even with this small example of two sentences, 11 features were created. The average English speaker knows around 8000 unique words. If each sentence were an 8000-sized vector sample in your dataframe, consisting mostly of 0's, it would be a poor use of memory.

SciPy implements sparse matrices as Python dictionaries: only the keys that have a value get stored, and everything else is assumed to be empty. You can always convert it back to a regular Python list by using the .toarray() method, but this converts it to a dense array, which might not be desirable due to memory usage reasons. To use your compressed, sparse, row matrix in Pandas, you're going to want to convert it to a Pandas <b><i>SparseDataFrame</i></b>
</p>

<h3>Bonus Topic: Functions</h3>

```python
def wordMatrixToDataFrame(wm, feat_names):
    # create an index for each row
    doc_names = ['Doc{:d}'.format(idx) for idx, _ in enumerate(wm)]
    df = pd.DataFrame(data=wm.toarray(), index=doc_names,
                      columns=feat_names)
    return(df)
```

```python
def CleanDataFrameV2(df):
    df.drop(["Opponent","Unnamed: 0_level_0"],inplace=True,axis=1)
    columns = ["Date","Tm","Opp","Result","MP"]
    columns = columns + [
        "FG_T",
        "FGA_T",
        "FG%_T",
        "2P_T",
        "2PA_T",
        "2P%_T",
        "3P_T",
        "3PA_T",
        "3P%_T",
        "FT_T",
        "FTA_T",
        "FT%_T",
        "PTS_T",
        "FG_O",
        "FGA_O",
        "FG%_O",
        "2P_O",
        "2PA_O",
        "2P%_O",
        "3P_O",
        "3PA_O",
        "3P%_O",
        "FT_O",
        "FTA_O",
        "FT%_O",
        "PTS_O"
    ]
    df.columns = columns
    df["Date"] = pd.to_datetime(df["Date"],errors='coerce') #To remove the headers within the table
    df.dropna(axis=0,inplace=True)
    return df
```

```python
#lets get the data and write it to a csv file so we don't have to do this again
url = ("https://www.basketball-reference.com/play-index/tgl_finder.cgi?"
       "request=1&player=&match=game&lg_id=NBA&year_min=2020&year_max=2020"
       "&team_id=&opp_id=&is_range=N&is_playoffs=N&round_id=&best_of=&team_seed=&"
       "opp_seed=&team_seed_cmp=eq&opp_seed_cmp=eq&game_num_type=team&game_num_min=&"
       "game_num_max=&game_month=&game_location=&game_result=&is_overtime=&c1stat=&"
       "c1comp=&c1val=&c2stat=&c2comp=&c2val=&c3stat=&c3comp=&c3val=&c4stat=&c4comp=&c4val=&"
       "order_by=pts&order_by_asc=&offset=")

start = 0
end = 1200
step = 100
for offset in range(start,end,step):
    try:
        link = url + str(offset)
        lst = pd.read_html(link)
        df = CleanDataFrameV2(lst[0])
        df.to_csv("Data/TeamLogs2020.csv",mode='a',index=False)
    except ValueError:
        print("Error")
```

```python
#finally we read back all the data into a new dataframe
teamlogs = pd.read_csv("Data/TeamLogs2020.csv")

#remove the headers
teamlogs["Date"] = pd.to_datetime(teamlogs["Date"],errors='coerce') #To remove the headers within the table
teamlogs.dropna(axis=0,inplace=True)
teamlogs.head(10)
```

```python
#Let's codified the wins (W = 1, L = 2)
teamlogs["ResultIndicator"] = teamlogs['Result'].str[:1]
teamlogs.loc[(teamlogs['ResultIndicator'] == 'W'),'ResultIndicator'] = 1
teamlogs.loc[(teamlogs['ResultIndicator'] == 'L'),'ResultIndicator'] = 0
teamlogs.head(5)
```

```python
#Let's get the Team Score and Opponent Score and the Differential
teamlogs[["TeamScore","OpponentScore"]] = teamlogs["Result"].str.extract('(\d+)-(\d+)')
teamlogs["TeamScore"] = pd.to_numeric(teamlogs["TeamScore"])
teamlogs["OpponentScore"] = pd.to_numeric(teamlogs["OpponentScore"])
teamlogs["Differential"] = teamlogs["TeamScore"] - teamlogs["OpponentScore"]
teamlogs.head(5)
```
